```python
N = 100

y = np.linspace(17,493,N)
x = np.linspace(-182,182,N)
print(x)
print(y)
```

```python
# matplotlib parameters
params = {'legend.fontsize' : 14, 
          'xtick.labelsize' : 14, 
          'ytick.labelsize' : 14, 
          'font.size' : 14}
plt.rcParams.update(params)
```

```python
fig, myPlot = plt.subplots()
myPlot.plot(x,y)
myPlot.set_xlabel('Random X')
myPlot.set_ylabel('Random Y')
myPlot.set_title('Plot')
myPlot.grid(True)
plt.show()
plt.savefig('plot_test.png')
```
